## Data Vizualisation

```python
## Plotly
import plotly.graph_objs as go
X, y = make_classification(n_features = 2, 
                           n_samples = 1000,
                           n_redundant = 0,
                           n_informative = 2,
                           random_state = 100)

fig = go.Figure()
fig.add_scatter(x = X[:, 0], 
                y = X[:, 1], 
                marker_size = 4, 
                mode = 'markers',
                marker_color = y)
fig.update_xaxes(showline = True,
                 mirror = True,
                 showgrid = False,
                 title = 'X_AXIS',
                 linecolor = 'black')
fig.update_yaxes(showline = True,
                 mirror = True,
                 showgrid = False,
                 title = 'Y_AXIS',
                 linecolor = 'black')
fig.update_layout(plot_bgcolor = '#F3C178')
```
