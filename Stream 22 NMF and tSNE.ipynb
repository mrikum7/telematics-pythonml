```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
data = load_digits()
```

```python
X = data.data
y = data.target
```

```python
X_ = StandardScaler().fit(X).transform(X)
pca = PCA(n_components = 2, whiten = True)
X_ = pca.fit(X_).transform(X_)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = X_[:, 0], 
                         y = X_[:, 1], 
                         mode = 'markers',
                         marker_color = y))
```

```python
X_t = TSNE(n_components = 2).fit_transform(X)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = X_t[:, 0], 
                         y = X_t[:, 1], 
                         mode = 'markers',
                         marker_color = y,  
                         marker_size = 4))
```
