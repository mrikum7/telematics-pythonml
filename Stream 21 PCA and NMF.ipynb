```python
from sklearn.datasets import make_classification, make_regression
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import plotly.graph_objs as go
```

```python
X, y = make_classification(n_classes = 2, 
                           n_features = 10)
```

```python
scalar = StandardScaler()
scalar.fit(X)
X_ = scalar.transform(X)
```

```python
pca = PCA(n_components = 3)
pca.fit(X_)
X__ = pca.transform(X_)
```

```python
fig = go.Figure(data = [go.Scatter3d(x = X_[:, 0], 
                                   y = X_[:, 1], 
                                   z = X_[:, 2], 
                                   mode = 'markers',
                                   marker_color = y, 
                                   marker_size = 4)])
fig.show()
```

```python
X, y = make_regression(n_features = 2, 
                       ) 
scalar = StandardScaler()
scalar.fit(X)
X_ = scalar.transform(X)
pca = PCA(n_components = 1, whiten = False)
pca.fit(X_)
X__ = pca.transform(X_)
fig = go.Figure(data = [go.Scatter(x = X_[:, 0], 
                                   y = X_[:, 1], 
                                    
                                   mode = 'markers',
                                   marker_color = y, 
                                   marker_size = 4)])
fig.show()
```

```python
pca = PCA(n_components = 1, whiten = True)
pca.fit(X_)
X__ = pca.transform(X_)
fig = go.Figure(data = [go.Scatter(x = X_[:, 0], 
                                   y = X_[:, 1], 
                                    
                                   mode = 'markers',
                                   marker_color = y, 
                                   marker_size = 4)])
fig.show()
```

```python
import numpy as np
```

```python
X = np.linspace(-10,10, 100) + 5 * np.random.random(100)
Y = np.linspace(-10,10, 100) + 5 * np.random.random(100)
```

```python
go.Figure(data = [go.Scatter(x = X, 
                             y = Y, 
                             mode = 'markers')])
```

```python
X = np.asarray((X, Y)).T
```

```python
scalar.fit(X)
X_ = scalar.transform(X)
pca = PCA(n_components = 2, whiten = True)
pca.fit(X_)
X__ = pca.transform(X_)
go.Figure(data = [go.Scatter(x = X__[:, 0], 
                             y = X__[:, 1], 
                             mode = 'markers')])
```

```python
import mglearn
from sklearn.model_selection import train_test_split
mglearn.plots.plot_pca_whitening()
```

```python
X, y = make_classification(n_features = 10,
                           n_samples = 1000,
                           n_classes = 2, 
                           n_informative = 2, 
                           n_redundant = 8)
```

```python
go.Figure(data = [go.Scatter(x = X[:, 0], 
                             y = X[:, 1], 
                             mode = 'markers', 
                             marker_color = y)])
```

```python
scalar.fit(X)
X = scalar.transform(X)
pca = PCA(n_components = 2, whiten = True)
pca.fit(X)
X = pca.transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 21)
clf = KNeighborsClassifier(n_neighbors = 3)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
