```python
from sklearn.datasets import make_blobs
```

```python
import plotly.graph_objs as go
```

```python
X, y = make_blobs(n_samples = 100, 
                  n_features = 3)
```

```python
y
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = X[:, 0], 
                           y = X[:, 1], 
                           z = X[:, 2], 
                           mode = 'markers', 
                           marker = {'color' : y, 
                                     'size' : y + 5}), 
             )
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn import naive_bayes
```

```python
clf = naive_bayes.GaussianNB()
```

```python
clf.fit(X, y)
```

```python
clf.score(X, y)
```

```python
fig = go.Figure()
x = ['A', 'B', 'C', 'D', 'E']
y = [4, 2, 5, 1, 7]
fig.add_trace(go.Scatter(x = x, 
                         y = y,
                         name = 'Scatter'), 
             )
fig.add_trace(go.Bar(x = x, 
                     y = y,
                     name = 'Bar Graph', 
                     marker = {'cmax' : 10, 
                               'cmin' : 0, 
                               'color' : y, 
                               'colorbar' : {'title' : 'Values'},
                               }))
fig.update_xaxes(title_text =  'Feature 1', 
                 title_font_family = 'Arial', 
                 title_font_size = 20, 
                 title_font_color = 'red', 
                 showline = True,
                 linewidth = 2, 
                 linecolor = 'red', 
                 mirror = True)
fig.update_yaxes(title = {'text' : 'Feature 2', 
                          'font' : {'family' : 'Raleway', 
                                    'size' : 10}})
fig.update_layout(plot_bgcolor = 'white')
fig.write_image('Test.png', scale = 4)
```
